# Artificial Neural Network

### Kütüphanelerin İçeri Aktarılması

```
import numpy as np
import pandas as pd
import tensorflow as tf
```

```
tf.__version__
```

## Part 1 - Veri Ön İşleme

### Veri Setinin İçeri Aktarılması

```
#Veri setini okuyarak bağımlı (X) ve bağımsız değişken (y) şekilinde ayırıyoruz.
dataset = pd.read_excel('Folds5x2_pp.xlsx')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
print(y)
```

### Veri Setinin Test ve Training Olarak Bölünmesi

```
#Veri setinin test ve train kümelerine bölüyoruz.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Part 2 - ANN'nin İnşa Edilmesi



### ANN'yi Başlatma

```
#Tensorflow Keras API'si kullanılarak bir ANN modeli oluşturulur.
ann = tf.keras.models.Sequential()
```

### Giriş katmanın ve ilk gizli katmanın eklenmesi

```
#6 nöron ve ReLU aktivasyon fonksiyonunu kullanan bir katman oluşturulur.
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### İkinci Gizli Katmanın Eklenmesi

```
#6 nöron ve ReLU aktivasyon fonksiyonunu kullanan ikinci gizli katman oluşturulur.
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Çıkış Katmanının Eklenmesi

```
#Genellikle regresyon kullanıldığı zaman activation fonksiyonu seçilmez.
ann.add(tf.keras.layers.Dense(units=1))
```

## Part 3 - ANN'nin Eğitilmesi

### ANN'nin Derlenmesi

```
#En popüler optimize edici olan adam kullanılmakta
#loss, modelin eğitim sırasında optimizasyon algoritması tarafından minimize edilecek kayıp fonksiyonunu belirler.
ann.compile(optimizer='adam', loss='mean_squared_error')
```

### ANN'yi Test Verileri Üzerinde Eğitme

```
#Oluşturulan ANN eğitilir.
ann.fit(X_train, y_train, batch_size = 32, epochs = 50)
```

### Test Verilerine Göre Sonuçların Tahmin Edilmesi

```
#İlk sütun tahmin verilerini belirtir.
#İkinci sütun gerçek verileri belirtir.
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## MAE - MSE - RMSE'nin Hesaplanması

```
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Ortalama Mutlak Hata (MAE) hesaplama
mae = mean_absolute_error(y_test, y_pred)
print("Ortalama Mutlak Hata (MAE):", mae)

# Ortalama Kare Hata (MSE) hesaplama
mse = mean_squared_error(y_test, y_pred)
print("Ortalama Kare Hata (MSE):", mse)

# Ortalama Kare Kök Hata (RMSE) hesaplama
rmse = np.sqrt(mse)
print("Ortalama Kare Kök Hata (RMSE):", rmse)
```
